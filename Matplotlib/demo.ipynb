```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
# Ploting Basic Graph
x=[5,10,15,20,25]
y=[6,12,18,24,30]
plt.plot(x,y)
plt.show()
```

```python
# Editing graph
plt.plot(x,y,color='red')
plt.show()
```

```python
# Add Marker and add color on Marker with size
x=[5,10,15,20,25]
y=[6,12,18,24,30]
plt.plot(x,y,color='yellow',marker='*',markeredgecolor='black',markersize=10)
plt.show()
```

```python
plt.plot(x,y,color='m',linestyle='dashdot',marker='*',markeredgecolor='black',markersize=10)
```

```python
# Sort form add color, marker, LineStyle
plt.plot(x,y,'b*--')
```

```python
# For heading of the graph
plt.plot(x,y,'r*--')
plt.title('Heading')
```

```python
# Handling Title font
plt.title('Heading',fontdict={'fontname':'Comic Sans Ms','fontsize':17})
```

```python
# Editing Graph using sort form
plt.plot(x,y,'r*--')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
```

```python
#Show graph catagory Using Legend
plt.plot(x,y,'g*--')
plt.legend(['single element'])
plt.show()
```

```python
# How to save graph
plt.plot(x,y,'g*--')
plt.savefig('First-graph.png',dpi=200)
```

```python
# How to show multiple Graph 
mobile=pd.read_csv('mobile.csv')
mobile.head()
```

```python
# Ploting Multiple Graph
plt.plot(mobile.year,mobile.india,'b.-',label='india')
plt.plot(mobile.year,mobile.china,'r.-',label='china')
plt.plot(mobile.year,mobile.japan,'g.-',label='japan')
plt.plot(mobile.year,mobile.america,'y.-',label='america')
plt.xlabel('Year')
plt.ylabel('Price')
plt.title('Mobile Price Year wise',fontdict={'fontname':'Comic Sans Ms','fontsize':15})
plt.legend()
plt.savefig('Mobile-price.png',dpi=300)
plt.show()
```

```python
# Histrogram Ploting

bunch=[0,20,40,60,80,100]
x=[10,20,30,40,50,60]
plt.hist(x,bins=bunch,color='black')
plt.xlabel('Bunch')
plt.ylabel('Data')
plt.legend(['Test'])
plt.show()
```

```python
#Piechart Ploting

y=np.array([52,35,47,50,40])
names=["ajay","sam","rahul","smita","shreya"]
plt.pie(y,labels=names)
plt.show()
```

```python
# Piechart Explode with add color and shadow property

nameexplode=[0.2,0,0,0,0]
namecolor=['b','hotpink','orange','m','black']
plt.pie(y,labels=names,explode=nameexplode,shadow=True,colors=namecolor)
plt.legend()
plt.show()
```

```python
# Ploting Multiple Boxplot 

test1=np.random.normal(200,50,400)
test2=np.random.normal(100,50,300)
test3=np.random.normal(150,100,300)
test4=np.random.normal(250,75,400)
data=[test1,test2,test3,test4]
plt.boxplot(data)
plt.show()
```

```python
# Ploting Scatter Plot

x=[1,2,3,4,5]
y=[5,6,7,8,9]
a=[7,4,3,8,2]
b=[9,5,3,6,1]
plt.scatter(x,y)
plt.scatter(a,b)
plt.show()
```
